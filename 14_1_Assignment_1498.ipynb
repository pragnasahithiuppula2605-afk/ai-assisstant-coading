Task 3 – Event Registration Form

```python
from IPython.display import display, HTML

html = """

<style>

form{
width:300px;
margin:auto;
font-family:Arial;
}

input,select{
width:100%;
padding:10px;
margin:10px 0;
}

button{
background:#007bff;
color:white;
padding:10px;
border:none;
}

</style>

<script>

function validateForm(){

let name=document.forms["form"]["name"].value

if(name==""){
alert("Name must be filled")
return false
}

return true
}

</script>

<h2 align=center>Event Registration</h2>

<form name="form" onsubmit="return validateForm()">

<label>Name</label>
<input type="text" name="name">

<label>Email</label>
<input type="email">

<label>Phone</label>
<input type="tel">

<label>Session</label>

<select>
<option>AI</option>
<option>Web Development</option>
<option>Cyber Security</option>
</select>

<button type="submit">Register</button>

</form>
"""

display(HTML(html))
```
